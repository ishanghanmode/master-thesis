```python
# 1 million Movielens dataset

ratings = pd.read_table('ratings.dat',sep='::',names=['user','movieid','rating','time'], engine='python')
users = pd.read_table('users.dat',sep='::',names=['user','gender','age','occupation','zip'], engine='python')
movies = pd.read_table('movies.dat',sep='::',names=['movieid','title','genre'], engine='python')
```

```python
movielens = pd.merge(ratings,movies)
```

```python
variance_1 = movielens_temp.groupby("movieid")['rating'].var(ddof=True)
```

```python
variance_2 = movielens_temp.groupby("movieid")['rating'].var(ddof=False)
```
