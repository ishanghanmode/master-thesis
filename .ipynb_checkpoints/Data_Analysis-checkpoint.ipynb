```python
#Load the MovieTweetings movie dataset 

movies_MT = pd.read_table('movies_MT.dat',sep='::',names=['imdbId','title','genre'], engine='python')
```

```python
movies_MT
```

```python
#Should have 7 characters in the imdbId field in Movielens Dataset

movielens_1M_final['imdbId'] = movielens_1M_final['imdbId'].apply(lambda x: '{0:0>7}'.format(x))
```

```python
#Should have 7 characters in the imdbId field in MovieTweetings Dataset

movies_MT['imdbId'] = movies_MT['imdbId'].apply(lambda x: '{0:0>7}'.format(x))
```

```python
movielens_1M_final_dataset = pd.merge (movielens_1M_final, movies_MT, on='imdbId')
```

```python
#To calculate the variance

ratings_1M_temp = ratings_1M.drop('user', 1)
```

```python
ratings_1M_temp
```

```python
ratings_1M_temp = ratings_1M_temp.sort_values('movieid', ascending=True)
```

```python
ratings_1M_temp 
```

```python
variance_1 = ratings_1M_temp.groupby("movieid")['rating'].var(ddof=True)
```

```python
variance_2 = ratings_1M_temp.groupby("movieid")['rating'].var(ddof=False)
```

```python
variance_3 = ratings_1M_temp.groupby("movieid")['rating'].var() #Default it is set as True
```

```python
variance_1
```

```python
variance_2
```
